```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Ler o Dataframe, como o arquivo CSV não possui títulos, utilizar o "header = None"
df = pd.read_csv('/content/Compras.csv', header = None)

#Transformando os NaN em 0
df.replace(np.nan,0, inplace = True)

df.head()
```

```python
# Definindo uma função para percorrer toda o dataframe linha a linha e retirar os 0, colocando cada linha em uma lista nova.
def removeAllOcurrencesOfValueInList(_list, value):
    return list(filter(lambda x: x != value, _list))
```

```python
list_all_transactions = []
#Index = número de cada linha, row = valor de cada linha (conteúdo)
# iterrows percorre cada linha do meu dataframe
for index, row in df.iterrows():
  #Converter a linha de dados do pandas para valores
    list_single_transaction = row.values.tolist()
  #Aplicando as funções para remover os 0.
    list_single_transaction = removeAllOcurrencesOfValueInList(list_single_transaction, 0)
  #Colocando na lista maior, cada lista criada a partir das linhas do DF.
    list_all_transactions.append(list_single_transaction)

list_all_transactions[0:10]
```

```python
from mlxtend.preprocessing import TransactionEncoder
#http://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/
# Função 'TransactionEncoder' utilizada para passar cada ítem como coluna e cada lista como linhas, e os valores sendo booleanos
# para ficar claro quais ítens estão em cada linha.

te = TransactionEncoder()
te_ary = te.fit(list_all_transactions).transform(list_all_transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

df.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
#min_support: se o item aparece em 1% das compras do dataset, ele pode ser levado em consideração, se menos que isso não.
#use_colnames = Usar os nomes das colunas.
frequent_itemsets = apriori(df, min_support = 0.01, use_colnames = True)
frequent_itemsets

#Ocorrência desses itens aparecerem nas compras do maior para o menor
frequent_itemsets.sort_values(by=['support'], ascending = False).head(10)
```

```python
from mlxtend.frequent_patterns import association_rules

# Pelo modelo de confiança, vamos filtrar apenas as compras que o item A e B estão juntos no mínimo 30% das vezes
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

#lift = < 1 significa que compra A, não compra B, lift = 1 é indiferente, > 2 = boa associação.
rules.sort_values(by=['lift'], ascending = False).drop(['antecedent support', 'consequent support', 'leverage', 'conviction'], axis=1)
```
